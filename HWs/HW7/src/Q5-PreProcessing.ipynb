## Question 5 - Part (A):
In this notebook we do pre-processing on the training data and the validation data. I use `Persian tweets emotional` dataset that you can download it from [[here]](https://www.kaggle.com/datasets/behdadkarimi/persian-tweets-emotional-dataset)

```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import random

pwd = os.getcwd()
```

```python
data_PATH = "../Data/Persian_tweets_emotional_dataset/anger.csv"
df = pd.read_csv(data_PATH)

print("Size of the dataset:", df.shape)
df
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize


nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Text cleaning function
def clean_text(text):
    text = re.sub(r"http\S+", "", text)  # Remove URLs
    text = re.sub(r"@\w+", "", text)  # Remove mentions
    text = re.sub(r"#\w+", "", text)  # Remove hashtags
    text = re.sub(r"[.,\n<>;:]", "", text)  # Remove specified characters
    text = re.sub(r"[^\w\s]", "", text)  # Remove punctuation
    return text
```

```python
# Apply text cleaning
df['cleaned_tweet'] = df['tweet'].apply(clean_text)
```

```python
print("Size of the dataset:", df.shape)
df['cleaned_tweet']
```

```python
# Tokenization
df['tokens'] = df['cleaned_tweet'].apply(word_tokenize)
```

```python
print("Size of the dataset:", df.shape)
df['tokens']
```

```python
# Removing stop words
stop_words = set(stopwords.words('english') + stopwords.words('persian'))
df['tokens'] = df['tokens'].apply(lambda x: [word for word in x if word not in stop_words])
```

```python
print("Size of the dataset:", df.shape)
df['tokens']
```

```python
# Shuffle tokens in each tweet
df['shuffled_tokens'] = df['tokens'].apply(lambda x: random.sample(x, len(x)))
df['shuffled_tokens']
```

```python
# Feature extraction: length of tweet and number of hashtags
df['tweet_length'] = df['cleaned_tweet'].apply(len)
df['num_hashtags'] = df['hashtags'].apply(lambda x: len(eval(x)) if pd.notna(x) else 0)
```

```python
# Encoding categorical variables (if needed)
df['sourceLabel_encoded'] = pd.factorize(df['sourceLabel'])[0]
df['emotion_encoded'] = pd.factorize(df['emotion'])[0]
```

```python
# Display the first few rows of the processed dataset
print("Processed dataset size:", df.shape)
df
```

```python
# Save the preprocessed data to a new CSV file
df.to_csv("../Data/Persian_tweets_emotional_dataset/processed_anger.csv", index=False)
```
