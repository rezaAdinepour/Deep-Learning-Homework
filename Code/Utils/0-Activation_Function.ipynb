```python
import torch
import matplotlib.pyplot as plt
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# create x points
x = torch.linspace(-5, 5, 100, requires_grad=True, device=device)
```

```python
y = torch.nn.functional.sigmoid(x)
y_g = y.backward(torch.ones_like(x))
```

```python
plt.figure(figsize=(12, 9))
plt.plot(x.cpu().detach().numpy(), y.cpu().detach().numpy(), color='blue', label='Sigmoid')
plt.plot(x.cpu().detach().numpy(), x.grad.cpu().detach().numpy(), color='red', label='Gradient')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()


plt.show()
```
